```python
url = "https://www.worldometers.info/coronavirus/#countries"
req = Request(url, headers={'User-Agent':"Mozilla/5.0"})

webpage = urlopen(req)
page_soup = soup(webpage, "html.parser")
page_soup 
```

```python
table = page_soup.findAll("table",{"id":"main_table_countries_yesterday"})
table
```

```python
containers = table[0].findAll("tr",{"style":""})
containers
```

```python
title = containers[0]
title
```

```python
for country in containers:
    country_data = []
    country_container = country.findAll("td")
    print(country_container)
```

```python
del containers[0]
all_data = []
clean = True
for country in containers:
    country_data = []
    country_container = country.findAll("td")
    
    if country_container[1].text == "China":
        continue
    for i in range(1, len(country_container)):
        final_feature = country_container[i].text
        if clean :
            if i != 1 and i != len(country_container)-1:
                final_feature = final_feature.replace(",","")
                if final_feature.find('+') != -1:
                    final_feature = final_feature.replace("+","")
                    final_feature = float(final_feature)
                elif final_feature.find("-") != -1:
                    final_feature = final_feature.replace("-","")
                    final_feature = float(final_feature)*-1
        if final_feature == 'N/A':
            final_feature = 0
        elif final_feature == "" or final_feature == " ":
            final_feature = -1
            
        country_data.append(final_feature)
        
    all_data.append(country_data)
```

```python
title
```

```python
all_headings = []
heads = title.findAll("th")
for i in range(1, len(heads)):
        final_feature = heads[i].text
        if final_feature.strip() == '1 Caseevery X ppl':
            break
        if final_feature.strip() != '':
             all_headings.append(final_feature.strip());

print(all_headings)
```

```python
import pandas as pd;
import numpy as np
```

```python
df = pd.DataFrame(all_data)
df.drop([15,16,17,18,19,20], inplace= True, axis = 1)
df.head()
```

```python
column_lables = all_headings
```

```python
column_lables
```

```python
df.columns = column_lables
df.head()
```

```python
for label in df.columns:
     if label != 'Country,Other' and label != 'Continent':
            df[label] = pd.to_numeric(df[label])
```

```python
df["%Inc Cases"] = df["NewCases"]/df["TotalCases"]*100
df["%Inc Deaths"] = df["NewDeaths"]/df["TotalDeaths"]*100
df["%Inc Recovered"] = df["NewRecovered"]/df["TotalRecovered"]*100
```
