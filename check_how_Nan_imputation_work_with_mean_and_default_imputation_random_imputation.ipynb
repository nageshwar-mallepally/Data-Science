```python
def impute_nan(df, variable, val):
    df[variable+"_mmm"] = df[variable].fillna(val)
    df[variable+"_random"] = df[variable].fillna(df[variable].mean())
    random_sample = df[variable].dropna().sample(df[variable].isnull().sum(), random_state = 0)
#     random_sample_index = df[df[variable].isnull()].index
    df.loc[df[variable].isnull(),variable+"_random"] = random_sample
```

```python
random_sample = df['age'].dropna().sample(df['age'].isnull().sum(), random_state = 1)
random_sample.isnull().sum()
```
